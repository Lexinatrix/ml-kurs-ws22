```python
import pandas as pd
data_raw = {'hours' : [1,1,1.5,2,2.5,2.5,2.5,3,3.5,3.5,3.5,4,4.5],
           'points' : [0.5,1.5,1,1.5,1.5,2.5,3,2,1.5,2,2.5,2,2.5]}
df = pd.DataFrame(data_raw)
```

Plot Anzeigen

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))
plt.scatter(df['hours'], df['points'])
plt.xlabel("Stunden (h)")
plt.ylabel("Punkte")
plt.show()
```
